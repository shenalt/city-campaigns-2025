```python
def smallOrLargeDonation(row):
    amt = row['AMNT']
    if amt <= 175:
        return 'Small'
    elif amt <= 999:
        return 'Medium'
    else:
        return 'Large'

copy_df['Contribution Size'] = copy_df.apply(smallOrLargeDonation, axis=1)
pd.reset_option('display.max_columns')
copy_df
```
